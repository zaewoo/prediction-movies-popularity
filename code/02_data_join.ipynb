### 3.3 배우 컬럼 전처리

```python
raw[raw['actor1_Nm'].isna()].sort_values('view', ascending=False).head(20)
```

```python
# 대다수의 배우가 없는 값이 애니메이션 장르이므로, 성우 이름이 많이 누락된 것으로 보인다.
# 애니메이션 장르의 경우 배우명을 수작업으로 네이버 영화에서 찾아 보았다.
# 해당 데이터가 많았기에 엑셀에서 작업 후, 판다스로 불러와 기존 데이터프레임에 덮어 씌우는 방식으로 진행

anime_act = pd.read_excel('./data/animationAltered.xlsx')
anime_act.head()
```

```python
# 기존 데이터프레임과 합치기

for idx, row in anime_act.iterrows():
    cd = str(row['movieCd'])
    nm = row['actor']
    try:
        i = raw.index[raw['movieCd']==cd].tolist()[0]
        raw.loc[i, 'actor1_Nm'] = nm
    except:
        continue
```

```python
# 추가적으로 수작업한 배우 명 불러오기
act_df = pd.read_excel('./data/actors_additional.xlsx')
act_df.head()
```

```python
# 기존 데이터프레임과 합치기

for idx, row in act_df.iterrows():
    cd = str(row['movieCd'])
    nm = row['actor1_Nm']
    try:
        i = raw.index[raw['movieCd']==cd].tolist()[0]
        raw.loc[i, 'actor1_Nm'] = nm
    except:
        continue
```

```python
# 나머지 결측치는 제거

raw = raw.dropna(subset='actor1_Nm', axis=0).reset_index(drop=True)
raw.info()
```

### 3.4 장르 단순화

```python
raw['genres_1'].value_counts()
```

```python
'''
1. 드라마/가족
2. 액션
3. 스릴러
5. 코미디
7. 멜로/로맨스
8. 공포(호러)
10. 다큐멘터리
11. 범죄
12. 애니메이션
'''
```

```python
# 장르 변환 또한 데이터 수가 많고 수작업이 필요했기에 엑셀에서 변경 후 기존 데이터프레임에 씌우기

genres = pd.read_excel('./data/genre_rename.xlsx')
```

```python
genres.head()
```

```python
# 기존 데이터프레임과 합치기

for idx, row in genres.iterrows():
    cd = str(row['movieCd'])
    g = row['genre']
    try:
        i = raw.index[raw['movieCd']==cd].tolist()[0]
        raw.loc[i, 'genres_1'] = g
    except:
        continue
```

```python
for idx, row in raw.iterrows():
    val = row['genres_1']

    if val == "드라마":
        raw.loc[idx, 'genres_1'] = 1
    elif val == "가족":
        raw.loc[idx, 'genres_1'] = 1
    elif val == "액션":
        raw.loc[idx, 'genres_1'] = 2
    elif val == "스릴러":
        raw.loc[idx, 'genres_1'] = 3
    elif val == "코미디":
        raw.loc[idx, 'genres_1'] = 5
    elif val == "멜로/로맨스":
        raw.loc[idx, 'genres_1'] = 7
    elif val == "다큐멘터리":
        raw.loc[idx, 'genres_1'] = 10
    elif val == "범죄":
        raw.loc[idx, 'genres_1'] = 11
    elif val == "애니메이션":
        raw.loc[idx, 'genres_1'] = 12
```

```python
raw.loc[4019, 'genres_1'] = 1
raw.loc[5911, 'genres_1'] = 2
raw.loc[6370, 'genres_1'] = 3
raw.loc[298, 'genres_1'] = 3
raw.loc[6331, 'genres_1'] = 3
raw.loc[4434, 'genres_1'] = 3
raw.loc[5742, 'genres_1'] = 3
raw.loc[6367, 'genres_1'] = 3
raw.loc[5751, 'genres_1'] = 1
raw.loc[6387, 'genres_1'] = 1
raw.loc[5243, 'genres_1'] = 10
raw.loc[5718, 'genres_1'] = 10
```

```python
# 나머지는 제거
remove = raw.index[(raw['genres_1'] == "기타") | (raw['genres_1'] == "뮤지컬")].tolist()
raw = raw.drop(remove, axis=0).reset_index(drop=True)
```

### 3.5 제작 국가명 단순화

```python
raw['nationNm'].value_counts()
```

```python
# 국가명을 1= 한국, 2= 미국, 3= 제3국으로 지정
for idx, row in raw.iterrows():
    val = row['nationNm']
    if val == "한국":
        raw.loc[idx, 'nationNm'] = 1
    elif val == "미국":
        raw.loc[idx, 'nationNm'] = 2
    else:
        raw.loc[idx, 'nationNm'] = 3
```

```python
raw['nationNm'].value_counts()
```

### 3.7 심의 등급 단순화

```python
# 청불 or NOT 청불로 단순화 작업
raw['watchGradeNm'].value_counts()
```

```python
import numpy as np
raw['is_adult'] = np.nan
```

```python
for idx, row in raw.iterrows():
    val = row['watchGradeNm']
    if (val == "청소년관람불가") | (val == "18세관람가"):
        raw.loc[idx, 'is_adult'] = 1
    else:
        raw.loc[idx, 'is_adult'] = 0
```
